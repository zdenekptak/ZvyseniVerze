```python
from stat import S_ISREG, ST_CTIME, ST_MODE, ST_MTIME, ST_GID
import os, sys, time
from datetime import datetime
import json
import subprocess
from configparser import ConfigParser
```

```python
def denSouboruVAdresari(cestaKAdresari):
    #  získame všechny soubory v adresáři s/ statistikami (fn jsou jednotlive soubory)
    seznamSouboruKeKontrole = [os.path.join(cestaKAdresari, fn) for fn in os.listdir(cestaKAdresari)]
    # 
    seznamStatCesta = [(os.stat(path), path) for path in seznamSouboruKeKontrole]
    # ponechat pouze běžné soubory, vložit datum vytvoření ((stat[ST_MTIME] - vlozi datum posledni zmeny)
    serazeneSoubory = ((stat[ST_MTIME], path) for stat, path in seznamStatCesta if S_ISREG(stat[ST_MODE]))    
    # smycka ktera vrati posledni modifikovny soubor
    for cdate, path in sorted(serazeneSoubory, reverse=True):
        datumSouboru = time.ctime(cdate)
        datumSouboruDen = [int(s) for s in datumSouboru.split() if s.isdigit()][0]
        return datumSouboruDen
    
def jsouDnesZmenyVSouborech(configCesty):
    
    # prace s json soubory
    with open(configCesty) as json_file:
        
        configCesty = json.load(json_file)
    
    # vytvoreni prazdnych slovniku pro zaznamenani posledni zmeny souboru
    listvsechnyZmenyOstre = []
    listvsechnyZmenyStare = []
    listvsechnyZmenyFreeze = []
    vsechnyZmenyOstre = {}
    vsechnyZmenyStare = {}
    vsechnyZmenyFreeze = {}
    # smycka pres vsechny adresare ktere kontrolujeme
    for zdroje, slovnikVyvoj in configCesty.items():
        if zdroje == 'Ostre':
            for vyvoj, cesta in slovnikVyvoj.items():               
                listvsechnyZmenyOstre.append((denSouboruVAdresari(cesta)))
                vsechnyZmenyOstre[vyvoj] = denSouboruVAdresari(cesta)                
        if zdroje == 'Stare':                                                                                        
            for vyvoj, cesta in slovnikVyvoj.items():                                            
                listvsechnyZmenyStare.append(denSouboruVAdresari(cesta))
                vsechnyZmenyStare[vyvoj] = denSouboruVAdresari(cesta)               
        if zdroje == 'Freeze':
            for vyvoj, cesta in slovnikVyvoj.items():                                   
                listvsechnyZmenyFreeze.append(denSouboruVAdresari(cesta))
                vsechnyZmenyFreeze[vyvoj] = denSouboruVAdresari(cesta)
                                             
    nowDate = datetime.now()  # current date and time
    datumDnesniDen = int(nowDate.strftime("%d"))

    konecneVysledky = {}
    if datumDnesniDen == max(listvsechnyZmenyOstre):
        konecneVysledky['ostre'] = True
    else:
        konecneVysledky['ostre'] = False
        
    if datumDnesniDen == max(listvsechnyZmenyStare):
        konecneVysledky['stare'] = True
    else:
        konecneVysledky['stare'] = False
        
    if datumDnesniDen == max(listvsechnyZmenyFreeze):
        konecneVysledky['freeze'] = True
    else:
        konecneVysledky['freeze'] = False
                        
    return konecneVysledky

def aktualniVerzeSVN(soubor):
    with open(soubor, 'r') as file:
        # načteme seznam řádků do data
        data = file.readlines()    
    # nyní vybereme 2. a 4. řádek
    aktualniVerzeSVN = []
    radky = [1,3]
    for i in radky:
        # nacteni konkretniho radku
        heoCisloVerze = data[i]
        # posledni 4 cisla
        posledni4Cisla = int(heoCisloVerze[-5:])
        posledni4Cisla = str(posledni4Cisla).zfill(4)
        aktualniVerzeSVN.append(posledni4Cisla)
    
    return aktualniVerzeSVN


def ziskaniCtyrCisli(soubor):
    with open(soubor, 'r') as file:
        # načteme seznam řádků do data
        data = file.readlines()    
    # nyní změnime 2. a 4. řádek, všineme si, že musíme přidat nový řádek /n znamená že odskočíme na další řádku
    radky = [1,3]
    for i in radky:
        # nacteni konkretniho radku
        heoCisloVerze = data[i]    
        # verze bez posledniho 4 cisli
        zacatekCislaVerze = heoCisloVerze[:-5]
        # posledni 4 cisla
        posledni4CislaInt = int(heoCisloVerze[-5:])
        posledni4Cisla = str(posledni4CislaInt).zfill(4)
        return posledni4Cisla  

def zvyseniVerze(soubor):    
    with open(soubor, 'r') as file:
        # načteme seznam řádků do data
        data = file.readlines()    
    # nyní změnime 2. a 4. řádek, všineme si, že musíme přidat nový řádek /n znamená že odskočíme na další řádku
    radky = [1,3]
    for i in radky:
        # nacteni konkretniho radku
        heoCisloVerze = data[i]    
        # verze bez posledniho 4 cisli
        zacatekCislaVerze = heoCisloVerze[:-5]
        # posledni 4 cisla
        posledni4Cisla = int(heoCisloVerze[-5:])
        zvyseny4Cisla = str(posledni4Cisla + 1).zfill(4)
        vyslednyStringVerze = zacatekCislaVerze + zvyseny4Cisla + '\n'
        data[i] = vyslednyStringVerze
    # and write everything back
    with open(soubor, 'w') as file:
        file.writelines(data)
        
def upraveniCommitSouboru(souborCommitSVN, message):
    
    with open(souborCommitSVN, 'r') as file:
        # načteme seznam řádků do data
        data = file.readlines()    
    # nyní změnime 14. řádek
    radky = [13]
    for i in radky:
        # nacteni konkretniho radku
        heoCisloVerze = data[i]
        cesta = '"c:\\Program Files\\TortoiseSVN\\bin\\svn.exe"'
        commit = 'commit C:\\HeO_vyroba_distribuce\\Zdroje\\Ostre\\asseco\\_Verze - kopie.inc -m'
        msg = f'"{message}"'
        x = cesta + ' ' + commit + ' ' + msg + '\n'
        data[i] = x
    # and write everything back
    with open(souborCommitSVN, 'w') as file:
        file.writelines(data) 
           
def updateFromSVN(): 
       subprocess.call([r'C:\Users\zdenek.ptak\Repository\CisloVerze\updateSvn.bat'])
        
def commitToSVN(): 
       subprocess.call([r'C:\Users\zdenek.ptak\Repository\CisloVerze\commitSvn.bat'])
        
def ziskaniVerzeZIni(iniSoubor):
    # instantiate
    config = ConfigParser()
    # načtení existujiciho souboru
    config.read(iniSoubor)
    # naštení hodnot z ini souboru
    heo2verzeOstre = config.get('HeO2_Ostre', 'verze')
    heo3verzeOstre = config.get('HeO3_Ostre', 'verze')
    heo2verzeStare = config.get('HeO2_Stare', 'verze')
    heo3verzeStare = config.get('HeO3_Stare', 'verze')
    heo2verzeFreeze = config.get('HeO2_Freeze', 'verze')
    heo3verzeFreeze = config.get('HeO3_Freeze', 'verze')
    return [heo2verzeOstre, heo3verzeOstre, heo2verzeStare, heo3verzeStare, heo2verzeFreeze, heo3verzeFreeze]

def ulozeniDoIni(iniSoubor, noveCisloVerze):
    config = ConfigParser()
    config.read(iniSoubor)
    # zmena existujicich hodnot
    config.set("HeO2_Ostre","verze", noveCisloVerze)
    config.set("HeO3_Ostre","verze", noveCisloVerze)
    
    with open(iniSoubor, 'w') as configfile:
        config.write(configfile)
    
def vyslednaZprava(cisloVerze):
    msg = f"Automatické zvýšení verze pro noční buildy: {cisloVerze}"
    return msg    
```

```python
souboryKeKontrole = "C:/Users/zdenek.ptak/Repository/CisloVerze/souborykekontrole.json"

souborOstre = "C:/HeO_vyroba_distribuce/Zdroje/Ostre/asseco/_Verze - kopie.inc"
iniSoubor = 'posledniUlozenaVerze.ini'
souborCommitSVN = "commitSvn.bat"


# updateFromSVN()
vysledkyPoslednichZmen = jsouDnesZmenyVSouborech(souboryKeKontrole)
aktualniSVN = aktualniVerzeSVN(souborOstre)
aktualniINI = ziskaniVerzeZIni(iniSoubor)

if aktualniSVN[0] == aktualniINI[0] and aktualniSVN[1] == aktualniINI[1]:
    print("Verze nebyla zvednuta manuálně")
    if vysledkyPoslednichZmen['ostre'] == True:
        print('Byly vloženy úpravy')        
        zvyseniVerze(souborOstre)
        noveCisloVerze = ziskaniCtyrCisli(souborOstre)
        ulozeniDoIni(iniSoubor, noveCisloVerze)
        message = vyslednaZprava(noveCisloVerze)
        upraveniCommitSouboru(souborCommitSVN, message)
        print(f"Verze byla zvysena na {noveCisloVerze}")
    else:
        print("Nebyly vloženy úpravy")        
else:
    print("Verze byla zvýšena")
```
